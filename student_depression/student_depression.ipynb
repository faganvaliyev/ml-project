```python
def plot_bar(y_value,x_value,column_name):

    plt.figsize=(6, 4)
    sns.countplot(x=y_value, hue=x_value, data=df, palette='muted')
    plt.title(f'Count of {column_name} by Depression Status')
    plt.xlabel(f'{df.groupby(y_value).size()}')
    plt.ylabel('Count')
    plt.xticks(rotation=0)
    plt.legend(title='Depression Status', labels=['Not Depressed', 'Depressed'])
    plt.tight_layout()
    plt.show()
```

```python
# Renaming specific columns for improved readability

df = df.rename(columns={
    'Have you ever had suicidal thoughts ?': 'suicidal_thoghts?',
    'Family History of Mental Illness' : 'family_mental_illness',
    'Academic Pressure':'academic_pressure',
    'Study Satisfaction' : 'study_satisfaction',
    'Sleep Duration' : 'sleep_duration',
    'Dietary Habits' : 'dietary_habits',
    'Financial Stress' : 'financial_stress'
})

# Converting column names to lowercase 
df.columns = df.columns.str.lower()
```

```python
category_mapping = {
    'Yes' : 1,
    'No' : 0,
    'Male' : 0,
    'Female' : 1
}

columns_to_map = ['gender','suicidal_thoghts?','family_mental_illness']

for col in columns_to_map:
    df[col] = df[col].map(category_mapping)
```

##### Encoding(OneHot Encoder)

```python
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OneHotEncoder

categorical_columns = df.select_dtypes(exclude='number').columns.to_list()

preprocessor = ColumnTransformer(
    transformers=[
        ('one_hot', OneHotEncoder(drop='first',sparse_output=False), categorical_columns)
    ],
    remainder='passthrough'
)

transformmed_array = preprocessor.fit_transform(df)

encoded_columns_names = preprocessor.named_transformers_['one_hot'].get_feature_names_out(categorical_columns)
all_column_names = list(encoded_columns_names) + [col for col in df.columns if col not in categorical_columns]

df_transformed = pd.DataFrame(transformmed_array,columns=all_column_names)
df_transformed
```

```python
df_transformed.isna().sum()
```

```python
df_transformed = df_transformed.fillna(0)
```

```python
df_transformed.isna().sum()
```

```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler

X,y = df_transformed.drop('depression', axis=1), df_transformed['depression']
```

```python
X_train_val,X_test,y_train_val,y_test = train_test_split(X,y,test_size=0.2,random_state=24)
X_train,X_val,y_train,y_val = train_test_split(X_train_val,y_train_val,test_size=0.2,random_state=41)
```

```python
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_val_scaled = scaler.transform(X_val)
X_test_scaled = scaler.transform(X_test)
```

```python
from sklearn.metrics import ConfusionMatrixDisplay,confusion_matrix
def conf_matrix(model, X, y,threshold=0.5):
    y_prob = model.predict_proba(X)[:, 1]  
    y_pred = (y_prob >= threshold).astype(int)

    cm = confusion_matrix(y, y_pred)
    cm_display = ConfusionMatrixDisplay(confusion_matrix = cm, display_labels = [0, 1])
    cm_display.plot()
```

```python
from sklearn.metrics import precision_score, recall_score, f1_score

def metrics(model, X, y,threshold=0.5):
    y_prob = model.predict_proba(X)[:, 1]  
    y_pred = (y_prob >= threshold).astype(int)
    
    precision = precision_score(y, y_pred)  
    recall = recall_score(y, y_pred)  
    f1 = f1_score(y, y_pred)  

    print(f"Precision is: {precision}, '\n'Recall is: {recall},'\n'F1_score is: {f1}")
    print()
```

```python
from sklearn.svm import SVC
from sklearn.ensemble import RandomForestClassifier
from sklearn.linear_model import LogisticRegression
from xgboost import XGBRFClassifier

log_reg = LogisticRegression(random_state=42,max_iter=10000)
rnd_clf = RandomForestClassifier(random_state=42)

models = [log_reg,rnd_clf]
for model in models:
    print(f'{model} and metrics')
    model.fit(X_train_scaled,y_train)
    metrics(model,X_val_scaled,y_val)
```

#### Randomized Search

```python
from sklearn.model_selection import RandomizedSearchCV

param_grid = {
    'C': [0.01, 0.1, 1, 5, 10],
    'solver': ['lbfgs', 'liblinear', 'saga'],
    'max_iter': [100, 300, 500, 1000, 5000, 10000]
}

random_search = RandomizedSearchCV(log_reg, param_grid, cv=5, error_score='raise',scoring='accuracy',n_jobs=-1,random_state=42,n_iter=10,)
random_search.fit(X_train_scaled, y_train)

# Best parameters
print("Best Parameters:", random_search.best_params_)
print("Best Cross-Validation Accuracy:", random_search.best_score_)
```

```python
from sklearn.model_selection import GridSearchCV

param_grid = {
    'C': [0.01, 0.1, 1, 5, 10],
    'solver': ['lbfgs', 'liblinear', 'saga'],
    'max_iter': [100, 300, 500, 1000, 5000, 10000]
}

grid_search = GridSearchCV(log_reg, param_grid, cv=5, error_score='raise',scoring='accuracy',n_jobs=-1)
grid_search.fit(X_train_scaled, y_train)

# Best parameters
print("Best Parameters:", grid_search.best_params_)
print("Best Cross-Validation Accuracy:", grid_search.best_score_)
```

```python
from sklearn.metrics import accuracy_score
best_model_log = grid_search.best_estimator_
y_pred = best_model_log.predict(X_val_scaled)
print("Validation Set Accuracy:", accuracy_score(y_val, y_pred))
```

```python
acc = accuracy_score(y_test,best_model_log.predict(X_test_scaled))
metrics(best_model_log,X_test_scaled,y_test)
print(acc)
```

```python
# Logistic regression without random or grid search is better
metrics(log_reg,X_test_scaled,y_test)
print(f' Accuracy score is: {accuracy_score(y_test,log_reg.predict(X_test_scaled))}')
```
