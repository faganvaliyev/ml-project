### **Converting and Renaming Columns**

```python
# Convert column names to lowercase 
df.columns = df.columns.str.lower()

# Rename specific columns for improved readability
column_renaming_map = {
    "photourl": "photo_url",
    "longname": "long_name",
    "playerurl": "player_url",
    "↓ova": "ova",
    "loan date end": "loan_date_end",
    "release clause": "release_clause",
    "heading accuracy": "heading_accuracy",
    "short passing": "short_passing",
    "fk accuracy": "fk_accuracy",
    "long passing": "long_passing",
    "ball control": "ball_control",
    "sprint speed": "sprint_speed",
    "shot power": "shot_power",
    "long shots": "long_shots",
    "standing tackle": "standing_tackle",
    "sliding tackle": "sliding_tackle",
    "gk diving": "gk_diving",
    "gk handling": "gk_handling",
    "gk kicking": "gk_kicking",
    "gk positioning": "gk_positioning",
    "gk reflexes": "gk_reflexes",
    "total stats": "total_stats",
    "base stats": "base_stats"

}


df = df.rename(columns=column_renaming_map)

# Display information about the DataFrame after renaming
df.info()
```

## Reordering Columns

```python
# Define desired column order
our_order = ['id', 'name', 'long_name', 'age', 'nationality', 'team & contract', 'positions', 'height', 'weight', 'foot', 'joined', 'value', 'wage', 'release_clause', 'ova']

# Select remaining columns not in the our order
remaining_columns = [col for col in df.columns if col not in our_order]

# Reorder columns
df = df[our_order + remaining_columns]

# Display information about the DataFrame after reordering
df.info()
```

### __Age Column__

```python
# Display age column
df["age"]
```

```python
# Add 4 years to 'age' column
df['age'] = df['age'] + 4

# Display a few rows
df['age']
```

### **Height Column**

```python
# Display the "height" column
df["height"]
```

```python
# Define a function to convert height from feet and inches to centimeters

def height_covert(height):
    feet,inches = height.split("'")
    feet = int(feet)
    inches = int(inches.replace('"',''))
    total_feet = feet + (inches * 0.08)
    height_in_cm = total_feet * 30.48
    return height_in_cm

# Apply the conversion function to the "height" column and convert to int
df["height_cm"] = df["height"].apply(height_covert).astype(int)

# Display the converted "height_cm" column
df["height_cm"]

# drop height
```

```python
df['height'] 
```

### **Weight Column**

```python
df['weight']
```

```python
# define a function to convert lbs to kg
def convert_weight(weight):
    weight = int(weight.replace("lbs",''))
    weight_in_kg = weight * 0.45
    return weight_in_kg

# apply function
df["weight_kg"] = df["weight"].apply(convert_weight).astype(int)

# Display the converted "weight_kg" column
df["weight_kg"]

# drop weight
```

## __Joined Column__

```python
# Display the first 5 values of the "joined" column
df["joined"].head()

# Convert the "joined" column to datetime format
df["joined"] = pd.to_datetime(df["joined"])

# Extract year, month, and day from the "joined" column
df["joined_year"] = df["joined"].dt.year
df["joined_month"] = df["joined"].dt.month
df["joined_day"] = df["joined"].dt.day

# Display the first 5 rows of the DataFrame after modifications
df.head()

# drop joined
```

### **Team & Contract Column**

```python
df['team & contract']
```

```python
# extract team names
df["team"] = df["team & contract"].str.extract(r'(\n\n\n\n.*\n)', expand=False).str.strip()

# Extract years range using a regex pattern
df["contract"] = df["team & contract"].str.extract(r'(\d{4} ~ \d{4})', expand=False)

# etract start and end date of contract
df['contract_start'] = df['contract'].str.slice(0,4)
df['contract_end'] = df['contract'].str.slice(-4)

# convert data to int and handle missing values
df['contract_start'] = pd.to_numeric(df['contract_start'],errors='coerce').fillna(0).astype(int)
df['contract_end'] = pd.to_numeric(df['contract_end'],errors='coerce').fillna(0).astype(int)

df.head()

# drop contract and team & contract
```

*   **Error handling:** Incorporated errors="coerce" in pd.to_numeric to handle potential conversion errors.
*   **fillna(0):** Used fillna(0) to replace missing values with 0 for contract start and end years.


### **Value, Wage and Release_Clause Columns**

```python
# Display the specified columns
df[["value", "wage", "release_clause"]]
```

```python
# Remove "€" from the columns and convert to numeric
for col in ["value", "wage", "release_clause"]:
    df[col] = df[col].str.replace("€", "", regex=False)
    df.loc[df[col].str.contains("M", na=False), col] = pd.to_numeric(df.loc[df[col].str.contains("M", na=False), col].str.replace("M", "")) * 1000000
    df.loc[df[col].str.contains("K", na=False), col] = pd.to_numeric(df.loc[df[col].str.contains("K", na=False), col].str.replace("K", "")) * 1000
    df[col] = df[col].astype(int)

# Rename columns
df = df.rename(columns={"value": "value_eur", "wage": "wage_eur", "release_clause": "release_clause_eur"})

# Display the updated "value_eur", "wage_eur"and "release_clause_eur" column
df[["value_eur", "wage_eur", "release_clause_eur"]]
```

### **Hits Column**

```python
# display sample from hist column
df["hits"].head(5)
```

```python
# remove characters
df["hits"] = df["hits"].str.replace("/n",'',regex=False)

# identy null values
df["hits"].isnull().sum()
```

```python
# Convert values containing "K" to integers (thousands)
df.loc[df["hits"].str.contains("K", na=False), "hits"] = pd.to_numeric(df.loc[df["hits"].str.contains("K", na=False), "hits"].str.replace("K", "", regex=False)) * 1000

# Fill missing values with 0 and convert to integers
df["hits"] = df["hits"].fillna(0).astype(int)

# Display the updated "hits" column
df["hits"].sample(5)
```

### **W/F, SM and IR Columns**

```python
# Display the specified columns
df[['w/f','sm','ir']]
```

```python
for col in ['w/f','sm','ir']:
    df[col] = df[col].str.replace("★","",regex=False).astype(int)

# display again
df[['w/f','sm','ir']]
```
